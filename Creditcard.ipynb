IMPORTING DEPENDENCIES

```python

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#LOADING THE DATASET TO A PANDAS DATASET
credit_card_data=pd.read_csv('/content/archive.zip')
```

```python
#PRINTING THE FIRST FIVE ROWS OF THE DATASET
credit_card_data.head()
```

```python
#PRINTING THE LAST FIVE ROWS OF THE DATASET
credit_card_data.tail()
```

```python
#GETTING THE INFORMATION REGARDING THE DATASET
credit_card_data.info()
```

```python
#DISTRIBUTING THE LEGIT AND FRAUDULENT DATA STRUCTURES
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
fraud.Amount.describe()
```

```python
#COMPARE THE VALUES FOR BOTH THE TRANSACTIONS
credit_card_data.groupby('Class').mean()
```

UNDER-SAMPLING

MAKE A DATASET WHICH CONSISTS OF EQUAL DISTRIBUTION OF LEGIT TRANSACTION AND FRAUDULENT TRANSACTION

```python
#RANDOMLY CHOOSES 492 SAMPLES OUT OF THE TOTAL LEG IT ENTRIES
legit_sample=legit.sample(n=492)
```

CONCATENATE TWO DATAFRAMES-->LEGIT_SAMPLE,FAUD

```python
#AXIS=0-[CONCATENATES ROWWISE] AND AXIS=1-[CONCATENATES COLUMNWISE]
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
#use the valuecout function on the new_dataset
#also apply it on the previous dataset which is credit_card_data
new_dataset['Class'].value_counts()
#valuecounts returns the no of entries and datatype
```

```python
new_dataset.groupby('Class').mean()
```

SPLITTING THE DATASET INTO FEATURES AND TARGETS

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

SPLIT THE DATA INTO TRAINING DATA AND TESTING DATA

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
#testsize =0.2 means 20% of the total dataset will be stored in the training dataset
#stratify=Y means that the label Y will be unigormly distributed in both training set and test set
#random=2 will make the splitting same
```

```python
print(X.shape , X_train.shape, X_test.shape)
```

MODEL TRAINING

LOGISTIC REGRESSION MODEL

```python
#LOADING ONE INSTANCE OF THE LOGISTIC REGRESSION MODEL INTO THE VARIABLE MODEL
model=LogisticRegression()
```

```python
#TRAIN THE LOGISTIC REGRESSION MODEL WITH THE TRAINING SET
model.fit(X_train,Y_train)
```

Now that we have already trained our model with the training set using the fit function we are ready to check the performance of the model

MODEL EVALUATION BASED ON ACCURACY SCORE

```python
#check accuracy for training set
training_set_accuracy=model.predict(X_train)
training_accuracy_score=accuracy_score(training_set_accuracy,Y_train)
print('Accuracy on training data :- ',training_accuracy_score)
```

```python
#check accuracy for test set
test_set_accuracy=model.predict(X_test)
test_set_accuracy_score=accuracy_score(test_set_accuracy,Y_test)
print('Accuracy on test data :- ',test_set_accuracy_score )
```

Accuracy on test set is important in order to evaluate its performance.But why do we calculate accuracy score for training set.

accuracy for trainig set is calculated for comparison. if the accuracy score for training set and test set are very different then there is a chance of overfitting or underfitting

if ACCURACY SCORE ON TRAINING SET > ACCURACY SCORE ON TEST SET THEN IT IS A CASE OF OVERFITTING

if ACCURACY SCORE ON TRAINING SET < ACCURACY SCORE ON TEST SET THEN IT IS A CASE OF UNDERFITTING
